```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
df=pd.read_csv('Churn_Modelling.csv')
df.head()
```

```python
df.shape
```

```python
#Dropping columns which are not required

df = df.drop(["RowNumber","CustomerId","Surname"],axis=1)
df
```

```python
df.columns
```

```python
#Converting Categorical labels to Numerical labels
#Gender and Geography are categorical columns

from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder

Label_Encoder = LabelEncoder()
df["Gender"] = Label_Encoder.fit_transform(df["Gender"])
df
```

```python
One_Encoder = OneHotEncoder()
df["Geography"]=One_Encoder.fit_transform(df["Geography"])
```

```python
One_Encoder = OneHotEncoder()
One_Encoded_final=One_Encoder.fit_transform(df[["Geography"]])
One_Encoded_final
```

```python
One_Encoder.get_feature_names_out()
```

```python
One_Encoder.__dir__()
```

```python
#Converting the values to array to see the data of OHE
One_Encoded_final = One_Encoded_final.toarray()
One_Encoded_final
```

```python
#Getting tht dataframe of the One_Encoded_final
df_One_Hot_Encoded = pd.DataFrame(data=One_Encoded_final,columns=One_Encoder.get_feature_names_out())
df_One_Hot_Encoded
```

```python
df_final = pd.concat([df.drop(["Geography"],axis=1),df_One_Hot_Encoded],axis=1)
df_final
```

```python
#Splitting the data into X and Y
X = df_final.drop(axis=1, columns=["Exited"])
Y= df_final["Exited"]
print(X.shape, Y.shape)
```

```python
#Train test split
X_train,X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.3,random_state=42)
print(X_train.shape,X_test.shape,Y_train.shape,Y_test.shape)
```

```python
#Standardization of numerical columns using StandardScaler
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
X_train
```

```python
#Storing the Encoders and Scaler for future use
import pickle

with open("Label_Encoder_Gender.pkl","wb") as file:
    pickle.dump(Label_Encoder,file)

with open ("OHE_Geographhy.pkl","wb") as file:
    pickle.dump(One_Encoder,file)

with open ("Standard_Scaler","wb") as file:
    pickle.dump(scaler,file)

```

ANN Implementation


```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.callbacks import EarlyStopping,TensorBoard
import datetime
```

```python
#Building the ANN model
model = Sequential(
    [
    Dense(units = 64, activation = "relu", input_dim = X_train.shape[1]),
    Dense(units=32,activation="relu"),
    Dense(units=1,activation="sigmoid")
     ]
)
```

```python
#Checking the parameters of the model

model.summary()
```

```python
#Adding the optimizer , loss function and metrics to the model
# For a Binary Classification problem we use Bianry Crossentropy as loss function and for multiclass classification we use Sparse Categorical Crossentropy
import tensorflow
optimizer = tensorflow.keras.optimizers.Adam(learning_rate=0.01)
loss_fucntion = tensorflow.keras.losses.BinaryCrossentropy()
metrics = "accuracy"

model.compile(optimizer= optimizer, loss=loss_fucntion, metrics=[metrics])

```

```python
#Training the model with EarlyStopping and TensorBoard Callbacks
from tensorflow.keras.callbacks import EarlyStopping, TensorBoard

call_backs = "log/fit/" + datetime.datetime.now().strftime("%Y%m%d-%H%M%S")
tensorboard_callback = TensorBoard(log_dir=call_backs, histogram_freq=1)

Early_Stopping_Callback = EarlyStopping(monitor="val_loss",patience=5,restore_best_weights=True)

```

```python
#Training the model with EarlyStopping and TensorBoard Callbacks and data
history = model.fit(X_train,Y_train,epochs=100, validation_data=(X_test,Y_test),callbacks=[Early_Stopping_Callback,tensorboard_callback])
```

```python
model.save("ANN_model.h5")
```
